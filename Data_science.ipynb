# Adding Titles

```python
import seaborn as sns
import matplotlib.pyplot as plt
flower=sns.load_dataset("iris")
flower
sns.lineplot(x="sepal_length", y="sepal_width", data=flower)
plt.title("plant of flowers")
plt.show()
```

# Adding of Limit

```python
import seaborn as sns
import matplotlib.pyplot as plt
flower=sns.load_dataset("iris")
flower
sns.lineplot(x="sepal_length", y="sepal_width", data=flower)
plt.title("plant of flowers")
plt.xlim(3)
plt.ylim(2)
plt.show()
```

#  Size of Figur

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_style(style=None, rc=None)
flower=sns.load_dataset("iris")
flower
plt.figure(figsize=(6,4))
sns.lineplot(x="sepal_length", y="sepal_width", data=flower)
plt.title("plant of flowers")
sns.set_style("darkgrid")
plt.show()
```

#  Horizontal Plot

```python
import seaborn as sns
from numpy import median
import matplotlib.pyplot as plt
boat=sns.load_dataset("titanic")
boat
sns.barplot(x="fare", y="class", hue="sex", data=boat, estimator=median)
plt.show()
```

```python
import seaborn as sns
from numpy import median
import matplotlib.pyplot as plt
boat=sns.load_dataset("titanic")
boat
sns.barplot(x="class", y="fare", data=boat, 
                linewidth=3, facecolor=(1,1,1,0),
                errcolor=".2",  edgecolor=".2")
plt.show()
```
